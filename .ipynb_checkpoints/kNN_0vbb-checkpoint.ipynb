```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
filename_b2b = '../data/b2b_Yb_1_0_preprocessed_SFFS.xlsx'
filename_one = '../data/one_Yb_1_0_preprocessed_SFFS.xlsx'
```

```python
b2b_D = pd.read_excel(filename_b2b)
one_D = pd.read_excel(filename_one)
b2b_D_shape = b2b_D.shape
one_D_shape = one_D.shape

b2b_D_resz = b2b_D[0:min(b2b_D_shape[0],one_D_shape[0])]
one_D_resz = one_D[0:min(b2b_D_shape[0],one_D_shape[0])]

resz_shape = b2b_D_resz.shape # they're the same shape now

print(resz_shape)
```

```python
X_feats_D = pd.concat([b2b_D_resz, one_D_resz], ignore_index = True)
X_feats = X_feats_D.to_numpy()
X_feats.shape[0]
```

```python
b2b_labls = np.ones( (resz_shape[0],), dtype=int) # 1 == found 0vbb
one_labls = np.zeros((resz_shape[0],), dtype=int) # 0 == found 1e
Y_labls = np.concatenate((b2b_labls,one_labls), axis=None)
Y_labls.size
```

```python
from sklearn.model_selection import train_test_split #sklearn.cross_validation is deprecated
x_train, x_test, y_train, y_test = train_test_split(X_feats, Y_labls, test_size = 0.3) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
T_Ns = np.zeros((100,), dtype=int)
F_Ps = np.zeros((100,), dtype=int)
F_Ns = np.zeros((100,), dtype=int)
T_Ps = np.zeros((100,), dtype=int)
nNeighbors = np.arange(1,101)
```

```python
for i in range(100):

    knn = KNeighborsClassifier(n_neighbors=nNeighbors[i])
    knn.fit(x_train, y_train)
    label_pred = knn.predict(x_test)
    #precisions[i] = metrics.accuracy_score(y_test, label_pred)
    #accuracies[i] = metrics.precision_score(y_test, label_pred)
    T_Ns[i], F_Ps[i], F_Ns[i], T_Ps[i] = metrics.confusion_matrix(y_test, label_pred).ravel()
    
    #print("accuracy score:")
    #print(metrics.accuracy_score(y_test, label_pred))
    #print(" ")
    #print("precision score:")
    #print(metrics.precision_score(y_test, label_pred))
```

```python
T_Ns = T_Ns / x_test.shape[0]
F_Ps = F_Ps / x_test.shape[0]
F_Ns = F_Ns / x_test.shape[0]
T_Ps = T_Ps / x_test.shape[0]
```

```python
#plt.plot(nNeighbors, T_Ps, label='true b2b')
plt.plot(nNeighbors[5:], F_Ps[5:], label='false b2b')
#plt.plot(nNeighbors, T_Ns, label='true 1e')
plt.plot(nNeighbors[5:], F_Ns[5:], label='false 1e')
plt.legend()
plt.xlabel('number of neighbors')
plt.title('kNN for Y,1,0,SFFS ; ' + str(X_feats.shape[0]) + ' b2b or 1e events')
```
